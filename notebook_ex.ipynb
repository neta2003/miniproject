```python
#part 2 - Q1
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,1,0.01) #making the timing of the sampeling
f = 5 #the frequency
sin_signal = np.sin(2*np.pi*x*f) #making the sinus signal
signal = np.diff(sin_signal) #the diff function findes the points where the slope changes direction
max_points = np.where((signal[:-1] > 0) & (signal[1:] < 0))[0]+1 #finding the points where there is change in the slope
                                                                    #from positive to negitive
min_points = np.where((signal[:-1] < 0) & (signal[1:] > 0))[0]+1 #finding the points where there is change in the slope
                                                                    #from negitive to positive
the_20_percent = 5 #reducing to 20%
downsampled_points = np.arange(0, len(sin_signal), the_20_percent) #the reduced sample points
downsampled_points = np.unique(np.concatenate([downsampled_points, max_points, min_points]))#making sure the max and min
                                                                                                #points are used
downsampled_signal = sin_signal[downsampled_points] #creating the downsampled points while keeping the original
                                                        #max and min points
downsampled_time = x[downsampled_points] #new time for the downsampled signal
plt.figure(figsize=(10,6)) #creating the figure and size
plt.plot(x, sin_signal, label="original", color="gray", alpha=0.7) #the original plot
plt.plot(downsampled_time, downsampled_signal, label="downsampled signal", color = "red", alpha = 0.7) #the new plot
plt.title("original and downsampled sinusidal signal")
plt.xlabel("time(seconds)")
plt.ylabel("amplitude")
plt.legend()
plt.grid(True)
plt.show()
#just in case the graph isn't clear enough proof it samples the max and min points of original
print("original maximum points", max_points)
print("original minimum points", min_points)
print("the downsampled points", downsampled_points)
```
